```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load Resnet Data data

train_resnet_loss = []
with open('../../tsn-pytorch/logs/train_resnet_loss.log') as f:
    for l in f:
        train_resnet_loss.append(float(l.strip()))
        
test_resnet_loss = []
with open('../../tsn-pytorch/logs/test_resnet_loss.log') as f:
    for l in f:
        test_resnet_loss.append(float(l.strip()))

train_resnet_perc1 = []
with open('../../tsn-pytorch/logs/resnet_train_prec1.log') as f:
    for l in f:
        train_resnet_perc1.append(float(l.strip()))

train_resnet_perc5 = []
with open('../../tsn-pytorch/logs/resnet_train_prec5.log') as f:
    for l in f:
        train_resnet_perc5.append(float(l.strip()))
```

```python
# Load bninception Data data

train_bninception_loss = []
with open('../../tsn-pytorch/logs/train_bninception_loss.log') as f:
    for l in f:
        train_bninception_loss.append(float(l.strip()))
        
test_bninception_loss = []
with open('../../tsn-pytorch/logs/test_bninception_loss.log') as f:
    for l in f:
        test_bninception_loss.append(float(l.strip()))

train_bninception_perc1 = []
with open('../../tsn-pytorch/logs/bninception_train_prec1.log') as f:
    for l in f:
        train_bninception_perc1.append(float(l.strip()))

train_bninception_perc5 = []
with open('../../tsn-pytorch/logs/bninception_train_prec5.log') as f:
    for l in f:
        train_bninception_perc5.append(float(l.strip()))
```

```python
def plot(x, y, xlabel, ylabel, title):
    plt.figure(figsize = [10,5])
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.plot(x, y)
    plt.show()
```

```python
plot(list(range(len(train_resnet_loss))), train_resnet_loss, 'iteration', 'train loss', 'Resnet Train Loss vs iteration')
```

```python
plot(list(range(len(test_resnet_loss))), test_resnet_loss, 'epoch (Freq=5)', 'test loss', 'Resnet test Loss vs epoch')
```

```python
plot(list(range(len(train_resnet_perc1))), train_resnet_perc1, 'iteration', 'train top@1', 'Resnet Top@1 Accuracy vs iteration')
```

```python
plot(list(range(len(train_resnet_perc5))), train_resnet_perc5, 'iteration', 'train top@5', 'Resnet Top@5 Accuracy vs iteration')
```

```python
plot(list(range(len(train_bninception_loss))), train_bninception_loss, 'iteration', 'train loss', 'BN Inception Train Loss vs iteration')
plot(list(range(len(test_bninception_loss))), test_bninception_loss, 'epoch (Freq=5)', 'test loss', 'BN Inception test Loss vs epoch')
plot(list(range(len(train_bninception_perc1))), train_bninception_perc1, 'iteration', 'train top@1', 'BN Inception Top@1 Accuracy vs iteration')
plot(list(range(len(train_bninception_perc5))), train_bninception_perc5, 'iteration', 'train top@5', 'BN Inception Top@5 Accuracy vs iteration')
```
